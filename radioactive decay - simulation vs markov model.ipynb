It states that you are given a large amount of some radioactive isotope, Compound A. It decays into a different isotope, Compound B. That, in turn, decays into a different compound, and so on. You are tasked with finding the proportion of each compound present in the sample after some number of seconds.

    p <- c(0.00007, # P(A -> B)
       0.0005,  # P(B -> C)
       0.00013, # P(C -> D)
       0.00022, # P(D -> S)
       0)       # S doesn't decay

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# probabilities of decay from current state to next state per second
p = np.array([0.00007, 0.0005, 0.00013, 0.00022, 0])
```

```python
# start with all the molecules in state A
molecules = np.array([1000, 0, 0, 0, 0])
```

```python
# time 
s = 5000
```

## Simulation

```python
# example of 100 trials of the first second
np.random.binomial(1000, p[0], 100)
```

```python
# create the simuluation
def sim_decay(molecules, s, p):
    n_molecules = len(molecules)
    
    # row for every time and a column for every isotope
    results = np.zeros(shape=(s+1, n_molecules))
    # initial condition
    results[0, ] = molecules
    
    for i in range(1, s+1):
        decayed = [np.random.binomial(molecule, prob) for molecule, prob in zip(results[i-1], p)]
        results[i] = results[i-1] + np.append([0], decayed[0:-1]) - decayed
    
    return results
    
```

```python
sim1 = sim_decay(molecules, 5000, p)
sim1
```

```python
df = pd.DataFrame(sim1)
df
```

```python
df.plot(kind='area', alpha=0.5)
```

```python
sim2 = sim_decay(molecules, 50000, p)
```

```python
df2 = pd.DataFrame(sim2)
```

```python
df2.plot(kind='area', alpha=0.5)
```

## Markov model

Create a transition matrix that can transform a vector of the number of particles in each state into a new vector 1 second forward in time.

```python
len(molecules)
```

```python
def make_transition(molecules, p):
    transition = np.zeros(shape=(len(molecules), len(molecules)))
    for i in range(len(molecules)):
        for j in range(len(molecules)):
            if i == j:
                transition[i,j] = 1 - p[i]
            elif j == i + 1:
                transition[i,j] = p[i]
    return transition
```

```python
transition = make_transition(molecules, p)
transition
```

```python
molecules.T.dot(transition)
```

```python
def markov_decay(molecules, s, p):
    t = make_transition(molecules, p)
    m = np.zeros(shape=(s+1, len(molecules)))
    
    m[0, ] = molecules
    for i in range(1, s+1):
        m[i, ] = m[i-1,].T.dot(t)
    
    return m
```

```python
markov1 = markov_decay(molecules, s, p)
```

```python
dfm1 = pd.DataFrame(markov1)
```

```python
dfm1.plot(kind="area", alpha=0.5)
```

```python
markov2 = markov_decay(molecules, s*10, p)
dfm2 = pd.DataFrame(markov2)
```

```python
dfm2.plot(kind="area", alpha=0.5, ylim=(0,1000))
```
